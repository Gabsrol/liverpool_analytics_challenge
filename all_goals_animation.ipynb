# Goals animation

```python
from moviepy import editor as mpy
from moviepy.video.io.bindings import mplfig_to_npimage

def draw_frame_x(df, t, fps, voronoi=False, show_info=True, goal_info='goal uninformed'):
    fig,ax,dfFrame = footyviz.draw_frame(df, t=t, fps=fps, show_info=show_info, goal_info=goal_info,display_time=True)
    if voronoi:
        fig, ax, dfFrame = footyviz.add_voronoi_to_fig(fig, ax, dfFrame)
    image = mplfig_to_npimage(fig)
    plt.close()
    return image    

def make_animation(df, fps=20, voronoi=False, show_info=True, goal_info='goal uninformed'):

    length=(df.index.max()+20)/fps
    clip = mpy.VideoClip(
        lambda x: draw_frame_x(df, t=x, fps=fps, voronoi=voronoi, show_info=show_info, goal_info=goal_info),
        duration=length-1).set_fps(fps)
    return clip
```

```python
full_clip = mpy.concatenate_videoclips(
    [make_animation(data.loc[play],goal_info=play) 
     for play in data.index.get_level_values('play').unique()])

full_clip.write_videofile("data_outputs/liverpool_analytics_all_goals.mp4")
full_clip.close()
```
